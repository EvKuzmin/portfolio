```python
import os
with open('matrix_columns.txt') as f_columns, open('matrix_values.txt') as f_values, open('temp.mtx','w') as f_temp:
    row = 0
    n_not_zero = 0
    max_col = 0
    for line_columns in f_columns:
        line_values = f_values.readline()
        line_columns = line_columns.replace('\n', '')
        line_values = line_values.replace('\n', '')
        columns = line_columns.split(',')
        values = line_values.split(',')
        for col, val in zip(columns, values):
            max_col = max(max_col, int(col))
            if val != '0.0':
                f_temp.write(str(row+1) + ' ' + str(int(col)+1) + ' ' + val + '\n')
                n_not_zero += 1
        row += 1
n_row = row
n_col = max_col+1
n_row, n_col, n_not_zero

with open('temp.mtx') as f_temp, open('matrix.mtx','w') as f_matrix:
    f_matrix.write('%%MatrixMarket matrix coordinate real general\n')
    f_matrix.write(str(n_row) + ' ' + str(n_col) + ' ' + str(n_not_zero) + '\n')
    for line in f_temp:
        f_matrix.write(line)
os.remove('temp.mtx')
```
